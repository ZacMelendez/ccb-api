```python
from dataclasses import dataclass
import pytablereader as ptr
from typing import List

with open('./ccb-docs.md', 'r') as file:
   raw = file.readlines()

with open('./ccb-docs.md', 'r') as file2:
   rawtext = file2.read()
   

@dataclass
class ServiceItem:
    srv: str
    required_props: dict = None
    opt_props: dict = None

def typeConverter(input):
    if 'string' in input or 'datetime' in input:
        return str
    elif 'int' in input or 'integer' in input:
        return int
    elif 'decimal' in input: 
        return float
    elif 'array' in input:
        return List
    elif input in ['boolean']:
        return bool
    else:
        print(input)

def MDToDict(input):
    data = []
    loader = ptr.MarkdownTableTextLoader(input)
    for table_data in loader.load():
        data.append(table_data.as_dataframe())

    data[0].reset_index()
    
    returnDict = {}

    for index, row in data[0].iterrows():
        returnDict[row['name']] = typeConverter(row['type'])
    return returnDict

lines = []

for i in raw:
    if i == '\n':
        continue
    else:
        lines.append(i.rstrip('\n'))
```

```python
services = []

for i in range(len(lines)):
    if lines[i] == '#### Service Name':
        tmp_item = ServiceItem(srv = (lines[i + 1]).rstrip('\n'))
        next_line = 1
        while lines[i + next_line] != "#### Service Name":
            if lines[i+next_line] == '#### Required Parameters':
                if lines[i+3].lower() == 'none':
                    pass
                else:
                    if lines[i+3][0] != '|':
                        continue
                    else:
                        tmp_table_string = ""
                        table_start = i+3
                        j = 0
                        while lines[table_start + j][0] == '|':
                            tmp_table_string += (lines[table_start + j] + '\n') 
                            j += 1
                        tmp_item.required_props = MDToDict(tmp_table_string)
            if lines[i+next_line] == '#### Optional Parameters':
                if lines[i+3].lower() == 'none':
                    pass
                else:
                    if lines[i+3][0] != '|':
                        continue
                    else:
                        tmp_table_string = ""
                        table_start = i+3
                        j = 0
                        while lines[table_start + j][0] == '|':
                            tmp_table_string += (lines[table_start + j] + '\n') 
                            j += 1
                        tmp_item.opt_props = MDToDict(tmp_table_string)
            next_line += 1
        services.append(tmp_item)
```
